```python
import librosa
import numpy as np
import librosa.display
import matplotlib.pyplot as plt
import scipy as sp
import IPython.display as ipd
```

```python
Audio_path = r'C:\Users\sande\OneDrive\Documents\GitHub\ASE_practice\DataSet\sound-of-pouring-liquid-into-a-glass-hd-263175.mp3'
ipd.Audio(Audio_path)
```

```python
signal, sr = librosa.load(Audio_path, sr=22050) # sr = sample rate
```

```python
plt.figure(figsize=(15, 5))
librosa.display.waveshow(signal, sr=sr, alpha=0.4)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

```python
ft = sp.fft.fft(signal)
magnitude = np.abs(ft)
frequency = np.linspace(0, sr, len(magnitude))
```

```python
plt.figure(figsize=(15, 5))
plt.plot(frequency, magnitude)
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.show()


plt.figure(figsize=(15, 5))
plt.plot(frequency[:5000], magnitude[:5000])
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.show()
```

```python
samples = range(len(signal))
t = librosa.samples_to_time(samples, sr=sr)

plt.figure(figsize=(15, 5))
plt.plot(t[1000:1400], signal[1000:1400])
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

```python
# create a sinsoidal wave
f = 523
phase_1 = 0
phase_2 = 0.5
sin = 0.5 * np.sin(2 * np.pi * f * t - phase_1)
sin_2 = 0.5 * np.sin(2 * np.pi * f * t - phase_2)
plt.figure(figsize=(15, 5))
plt.plot(t[1000:1400], sin[1000:1400])
plt.plot(t[1000:1400], sin_2[1000:1400])
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```
